```python
import networkx as nx
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def nodes_from_csv(path):
    df = pd.read_csv(path, index_col=0)
    nodes = []
    
    for ix , row in df.iterrows():
        d = dict(row)
        node = d.pop('node')
        nodes.append((node, {k:v for k,v in d.items() if v}))
        
    return nodes
```

```python
#Import Vertices of the graph
nodes = nodes_from_csv('jena_graph.csv')
```

```python
#Import Edges of the graph
fname = os.path.join("adj_matrix.csv")

adj_matrix = pd.read_csv(fname)
```

```python
# graph construction:
#H = nx.from_pandas_edgelist(adj_matrix, create_using=nx.MultiDiGraph)
H = nx.from_pandas_edgelist(adj_matrix, create_using=nx.Graph)
H.update(nodes=nodes)
```
